```python
# simple list url ETL
import re

DOMAIN = "http://www.appledaily.com.tw"


# Parsing the HTML
urls = []
for x in range(1,11):
    print("[DEBUG] x is %s"%x)
    with open('./appledaily_%s.html'%x) as f:
        doc = f.read()
        
    url_with_hosts = [DOMAIN + d for d in re.findall('href="(/realtimenews/article/.*/.*/.*/.+)" target', doc)]
    
    urls.extend(url_with_hosts)
```

```python
# serialization, deserialization
import json

with open('./urls.json', 'w') as f:
    f.write(json.dumps(urls, ensure_ascii=False, indent=4))
```
